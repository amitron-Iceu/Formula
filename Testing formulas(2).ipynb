```python
#Make dataframe with only over-expression gene-sets
gsup = gs1.filter(like='UP', axis=1)
gsup = gsup.transpose()
#remove the _UP suffix
gsup.index = gsup.index.str.replace(r'_UP$','')
gsup = gsup.transpose()
gsup

#Make dataframe with only under-expressed gene-sets
gsdn = gs1.filter(like='DN', axis=1)
#multiply everything by -1 since we will be using the negative sign to represent under-expression
gsdn = gsdn*-1
gsdn = gsdn.transpose()
#remove the _DN suffix 
gsdn.index = gsdn.index.str.replace(r'_DN$','')
gsdn = gsdn.transpose()

#Add the values together; you will get a dataframe with +1's and -1's for each gene-set
# (+1) = over-expression and (-1) = under-expression
combgs = gsup+gsdn
combgs = combgs.transpose()
size_of_gs = combgs.abs().sum(axis=1)
#sum of absolute values for each geneset and make it a new column 
combgs["sum"] = combgs.abs().sum(axis=1)
#divide each value by the sum of the row it is in 
gszeros = combgs.loc[:,"G1":"G20"].div(combgs["sum"], axis=0)
gszeros = gszeros.transpose()

#Replace the zeros with NaNs since we can't claim that a gene is normally expressed if 
#there is no data for it
gs = gszeros.replace(0,np.nan,inplace=False)


# Will need a list of the gene-sets later without the _UP or _DN 
list_of_gs = ['COLD', 'HOT', 'FOOD','G5KO', 'G10PLUS', 'CHEMX', 'MUSH', 'G11KO']
```

```python
size_of_gs = pd.DataFrame(data = size_of_gs)
size_of_gs.columns = ['size']
size_of_gs = size_of_gs.sort_values(by='size')
```

```python
#Filter all down regulated genes
dn = tdf.filter(like='DN', axis=0)
#representing the under-expressed genes as having negative values. 
dn = dn*-1
dn = dn.reset_index()
dn['Gene'] = dn['Gene'].str.replace(r'_DN$','')
dn = dn.set_index('Gene')
#dn
```

```python
#Filter all up regulated genes
up = tdf.filter(like='UP', axis=0)
up = up.reset_index()
up['Gene'] = up['Gene'].str.replace(r'_UP$','')
up = up.set_index('Gene')
#up
```

```python
#add up the values of the positive and negative values for each gene 
#This get's rid of the "how small is too small" problem we were having before!
ldaP = dn+up
ldaP = ldaP.reset_index()

# ========================THIS STEP NOT RELEVANT, ONLY USING FOR THE MODEL ORGANISM ===================
zzz = {
    'Gene': list(range(11,21)),
    'Topic_1':np.zeros(10),
    'Topic_2':np.zeros(10),
    'Topic_3':np.zeros(10),
    'Topic_4':np.zeros(10)
}
zeros = pd.DataFrame(data=zzz)
zeros = zeros.loc[:,]
zeros['Gene'] = 'G' + zeros['Gene'].astype(str)

ldaP = pd.concat([ldaP,zeros], axis=0)
ldaP = ldaP.set_index('Gene')
#ldaP
```

```python
#create a dataframe made up of the first topic only 
temp = pd.DataFrame(data=ldaP['Topic_1'])
#make a new dataframe where the topic column is repeated n times 
#where n is the number of gene-sets
t1 = pd.concat([temp]*gs.shape[1],axis=1)
t1
```

```python
phase1 = []
phase2 = []
phase3 = []
phase4 = []
phase5 = []
phase6 = []
phase7 = []
phase8 = []
phase9 = []
distances = []

i=0
for i in range(len(topic_list)):
    phase1.append(pd.DataFrame(data=topic_list[i].values*gszeros.abs().values))
    phase2.append(phase1[i].transpose())
    phase2[i]["sum"] = phase2[i].abs().sum(axis=1)
    phase3.append(phase2[i].iloc[:,0:20].div(phase2[i]["sum"], axis=0))
    phase4.append(phase3[i].transpose())
    phase5.append(pd.DataFrame(data=gs.values-phase4[i].values).abs())
    phase5[i].columns = list_of_gs
    phase5[i].insert(2,'Gene',list(range(1,21)))
    phase5[i]['Gene'] = 'G' + phase5[i]['Gene'].astype(str)
    phase6.append(phase5[i].set_index('Gene'))
    phase7.append(phase6[i].dropna(axis=1,how='all'))
    phase8.append(pd.DataFrame(data = {'Distance':phase7[i].sum(axis=0)}))
#    phase9.append((2-phase8[i])/2)
    distances.append(phase8[i].sort_values(by='Distance'))


i = 0
for i in range(len(topic_list)):
    print('For Topic ', i+1, 'here are the gene-set distances:')
    print(distances[i])
```
